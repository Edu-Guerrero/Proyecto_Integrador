```python
# prompt: open my csv and save the contents of row 2 in a single txt file called combined_text.txt

import pandas as pd

# Assuming your CSV file is named 'your_csv_file.csv'
df = pd.read_csv('/content/englishpreprocessed_texts.csv')

# Extract the contents of the second column, starting from the second row
second_column_contents = df.iloc[1:, 1].tolist()  # .iloc[1:, 1] means row 2 onward, second column

# Combine the text with newline characters to write to a single file
combined_text = '\n'.join(map(str, second_column_contents))

# Save the combined text to a file
with open('combined_text.txt', 'w') as f:
    f.write(combined_text)
```

# With all the courses as a single doc #

```python
# prompt: save my combined_text.txt in a variable called doc

with open('combined_text.txt', 'r') as f:
  doc = f.read()
```

```python

# Remove numeric characters from the doc
doc = ''.join([i for i in doc if not i.isdigit()])
```

```python
import os
# Ensure the directory exists
os.makedirs("all_in_one", exist_ok=True)
```

## Distilbert##

```python
model_distilbert = KeyBERT(model="distilbert-base-nli-mean-tokens")
```

### Most simple One ###

```python
keywords = model_distilbert.extract_keywords(
    doc,
    top_n = 50,
    keyphrase_ngram_range=(1, 1),
    stop_words='english'
)

# Save keywords to a text file
with open("all_in_one/keywords_dist_simple_one.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Keyword", "Score"])  # Header
        writer.writerows(keywords)  # Write each keyword and its score as rows
```

# With all courses as different docs #

```python
# prompt: read my csv file and create an array to save each row in the second column

import pandas as pd

def read_csv_second_column(file_path):
  try:
    df = pd.read_csv(file_path)
    second_column_array = df.iloc[:, 1].tolist()  # Assuming the second column is at index 1
    return second_column_array
  except FileNotFoundError:
    print(f"File not found: {file_path}")
    return []
  except Exception as e:
    print(f"An error occurred: {e}")
    return []


# Example usage:
file_path = "/content/englishpreprocessed_texts.csv"  # Replace with your file path
second_column_data = read_csv_second_column(file_path)
```

## Distilbert ##

```python
model_distilbert = KeyBERT(model="distilbert-base-nli-mean-tokens")
```

### Most simple one ###

```python
import os
# Ensure the directory exists
os.makedirs("simple_one", exist_ok=True)
```

```python
for idx, doc in enumerate(second_column_data):
  keywords = model_distilbert.extract_keywords(
      doc,
      top_n = 5,
      keyphrase_ngram_range=(1, 1),
      stop_words='english'
  )

  # Save keywords to a text file
  with open(f"simple_one/keywords_dist_simple_one_{idx}.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Keyword", "Score"])  # Header
        writer.writerows(keywords)  # Write each keyword and its score as rows
print("Keywords saved to individual text files in the 'simple_one' directory")
```
